```python
from verstack import NaNImputer
ckd.isnull().sum().sum()

imputer=NaNImputer()
ckd_imputed = imputer.impute(ckd)
ckd_imputed.isnull().sum().sum()
```

```python
ckd_imputed = ckd_imputed.replace('\t', '', regex=True)
```

```python
ckd_imputed = ckd_imputed[~ckd_imputed.isin(['?']).any(axis=1)]
```

```python
ckd_imputed.drop_duplicates()
ckd_cleaned = ckd_imputed.drop('id', axis=1)
```

```python
ckd_cleaned.head()
```

```python
ckd_cleaned.head()
```

```python
for col in cat:
    print(f"{col} has {ckd_cleaned[col].nunique()}")
```

```python
ckd_cleaned['classification'].value_counts()
ckd_cleaned['classification'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for col in cat:
    ckd_cleaned[col] = le.fit_transform(ckd_cleaned[col].astype(str))
categorical_features_to_encode = ['htn', 'dm']
for col in categorical_features_to_encode:
    le = LabelEncoder() 
    ckd_cleaned[col] = le.fit_transform(ckd_cleaned[col].astype(str))


```

```python
print(ckd_cleaned['classification'].value_counts())
```

```python
ckd_cleaned.head()
```

```python
X = ckd_cleaned.drop('classification',axis=1)
Y = ckd_cleaned['classification']
```

```python
X
```

```python
Y
```

```python
Y.value_counts()
```

```python
from sklearn.model_selection import train_test_split


X_train,X_test, Y_train, Y_test =  train_test_split(X,Y, test_size = 0.1, random_state = 0)

print(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
important_features = ['sc', 'al', 'sg', 'hemo', 'htn', 'dm']
X_train_important = X_train[important_features]
X_test_important = X_test[important_features].reindex(columns=X_train.columns)

lr.fit(X_train_important, Y_train)
```

```python
print("Training Features:", X_train_important.columns)
print("Test Features:", X_test_important.columns)
```

```python
X_test_important = X_test_important.reindex(columns=X_train_important.columns)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

lr_acc = accuracy_score(Y_test, lr.predict(X_test_important))
print(f"Training Accuracy of LR is {accuracy_score(Y_train, lr.predict(X_train_important))}")
print(f"Testing Accuracy of LR is {accuracy_score(Y_test, lr.predict(X_test_important))}")

print(f"Confusion Matrix of LR is \n {confusion_matrix(Y_test, lr.predict(X_test_important))}\n")
print(f"Classification Report of LR is \n{classification_report(Y_test, lr.predict(X_test_important))}")
```
